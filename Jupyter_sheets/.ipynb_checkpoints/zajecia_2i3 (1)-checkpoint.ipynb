Możemy także sprawdzić jak długa jest lista.

<h2 style="color: darkgreen">Pętla</h2>
<p>Powtarzanie pewnego ciągu instrukcji (fragmentu kodu) <b>skończoną</b> ilość razy.<br> Np.<br> - pytaj o kolejną liczbę dopóki nie trafisz na zero <br> - powtarzaj działanie 7 razy<br> - wykonaj określoną instrukcję dla wszystkich elementów listy</p>

W pythonie występują dwa rodzaje pętli:

Pętla <b>for</b>:

```python
liczby = [2, 3, 5, 7]
for liczba in liczby:
    print(liczba)
```

Pętla <b>while</b>:

```python
licznik = 0
while licznik < 5:
    print(licznik)
    licznik += 1
```

<h3 style="color: darkblue">Pętla for:</h3>

- dla kolejnych znaków w stringu

```python
for litera in 'Python':
    print('Bieżąca litera: ', litera)
```

- dla elementów list

```python
owoce = ['banan', 'jabłko', 'gruszka']
for owoc in owoce:
    print('Owoc: ', owoc)
```

- dla zakresu liczb

```python
for num in range(10, 20,2): #od 10 do 19
    print(num)
    
for num in range(10): # od 0 do 9
    print(num)
```

<b>range()</b> zachowuje się podobnie jak zakresy elementów listy - składa się z trzech składowych.

range(kiedy_stop)

range(start, kiedy_stop)

range(start, kiedy_stop, krok)

```python
print('range(kiedy_stop)')
for i in range(5):
    print(i)
    
print('range(start, kiedy_stop)')
for i in range(2, 10):
    print(i)
    
print('range(start, kiedy_stop, krok)')
for i in range(0, -10, -2):
    print(i)
```

<p>Policz średnie wynagrodzenie za rok 2017. Program powinien spytać o wynagrodzenie z każdego miesiąca, a następnie policzyć i wypisać średnią.<br> średnie = suma / liczba_miesięcy<br>Napisz zadanie w dwóch wersjach, z wykorzystaniem pętli <b>for</b> i <b>while</b></p>

```python
miesiące = ['styczeń', 'luty', 'marzec', 'kwiecień', 'maj', 'czerwiec', 'lipiec', 'sierpień', 'wrzesień', 'październik', 'listopad', 'grudzień']
wynagrodzenie = []
suma = 0
i = 0

for items in miesiące:
    wynagrodzenie.append(int(input("Podaj wynagrodzenie z miesiąca {}\n".format(items))))
    suma += wynagrodzenie[i]
    i += 1
        
average = suma/len(miesiące)
print(average)
```

```python
miesiące = ['styczeń', 'luty', 'marzec', 'kwiecień', 'maj', 'czerwiec', 'lipiec', 'sierpień', 'wrzesień', 'październik', 'listopad', 'grudzień']
suma = 0
i = 0

while (i < len(miesiące)):
    suma += int(input("Podaj wynagrodzenie za miesiąc {}\n".format(miesiące[i])))
    i += 1
    
average = suma/len(miesiące)
print("Średnie wynagrodzenie w tym roku wynosi {}".format(average))
```

<h3 style="color: darkblue">Break, continue</h3>

<b> Break </b> pozwala na przerwanie pętli.<br>
<b> Continue </b> przechodzi do następnej iteracji pętli.

```python
while True:
    liczba = int(input("Podaj liczbę: "))
    print(liczba)
    if liczba == 0:
        break
```

```python
for num in range(2, 10):
    if num % 2 == 0:
        print("Liczba parzysta", num)
        continue
    print("Liczba nieparzysta", num)
```

```python
dni_tygodnia = ("poniedziałek", "wtorek", "środa", "czwartek", "piątek", "sobota", "niedziela")
userInput_weekDay = input("Jaki mamy dzień tygodnia? \n").lower()
dayNumber = 1

for items in dni_tygodnia:
    if userInput_weekDay == items:
        if dayNumber > dni_tygodnia.index("piątek"):
            print("Jest już weekend!")
        else:
            print("Do weekendu dostało {} dni.".format((5-dayNumber)))
    dayNumber += 1
            
```

```python
#alt
dni_tygodnia = ("poniedziałek", "wtorek", "środa", "czwartek", "piątek", "sobota", "niedziela")

dzien = input("Jaki jest dzien tygodnia? ")

indeks = dni_tygodnia.index(dzien)

ile_dni_zostalo = dni_tygodnia.index("piątek") - indeks
if ile_dni_zostalo <= 0:
    print("Jest weekend")
else:
    print("Zostalo "+str(ile_dni_zostalo)+ " dni")
```

<h2 style="color: darkgreen">Słowniki</h2>
<p>Słownik jest strukturą danych podobną do listy, ale pracuje się na niem w oparciu o klucze, a nie indeksy. Klucze i wartości mogą być obiektami dowolnego typu (napisy, liczby, listy itp.</p>

```python
#słownik = {klucz1:wartość, klucz2:wartość, klucz3:wartość}

przyjaciele = ['Michał', 'Ola', 'Bartek', 'Tomek']
miasta_przyjaciół = {'Michał':'Poznań',
                    'Ola':'Kraków',
                    'Bartek':'Berlin',
                    'Tomek':'Poznań',
                    (1,2,3):[1,2,3]}
print(miasta_przyjaciół['Tomek'])
print( miasta_przyjaciół[(1,2,3)][1]     )
```

WAŻNE!!
Klucze muszą być unikalne.
Wartości mogą być takie same dla kilku kluczy.
W słownikach nie ma określonej kolejności elementów, są one nieposortowane - przeciwieństwo listy, gdzie elementy mają konkretną kolejność.

<b>Pobieranie informacji ze słownika</b><br>
Z jakiego miasta jest Ola?

```python
miasto_oli = miasta_przyjaciół['Ola']
print(miasto_oli)
```

```python
miasta_przyjaciół['Ola'] = 'Kijów'
nowe_miasto_oli = miasta_przyjaciół['Ola']
print(nowe_miasto_oli)
```

<b>Dodawanie nowych elementów do słownika</b><br>
Poznałam Grzegorza. Gdzie mieszka Grzegorz?

```python
miasta_przyjaciół['Grzegorz'] = 'Poznań'
miasto_grzegorza = miasta_przyjaciół['Grzegorz']
print(miasto_grzegorza)
```

<h2 style="color: darkgreen">Funkcje</h2>
<p>Funkcje to wydzielone fragmenty kodu. Kod funkcji jest definiowany jednokrotnie, natomiast może być używany wielokrotnie. Funkcje mogą przyjmować argumenty i zwracać wartości.</p>

```python
def moja_funkcja(): # definicja funkcji
    print("Hello world")
```

```python
moja_funkcja() # wywolanie funkcji
moja_funkcja()
moja_funkcja()
```

```python
lista = [1,2,34]
a = len(lista)
print(a)

zmienna = moja_funkcja()
print(zmienna) # ta funkcja nic nie zwraca
```

```python
def moja_funkcja(): # definicja funkcji
    return "Hello world"
```

```python
zmienna = moja_funkcja()
print(zmienna) # ta funkcja zwraca ciąg znaków
"Hello world" == moja_funkcja()
```

```python
def suma(x,y): # funkcje moga przyjmowac argumenty
    s = x+y
    return s
```

```python
suma(3,4)
```

```python
suma("ala",4) # funkcje nie sprawdzają typów
```

```python
def potega(x, y=2): # argumenty moga przyjmowac wartosci domyslne
    return x**y
```

```python
potega(2)
```

```python
potega(2,3)
```

```python
def potega(x=3, y=2):
    a = (x**y, x, y)
    return a
```

```python
p = potega()
print(p)
```

```python
potega()
```

```python
potega(4)
```

```python
# jak powiedziec funkcji ze argument, ktory podajemy ma byc zapisany do argumentu y?
potega(y=4)
```

```python
potega(y=4, x=2) # mozemy wtedy uzywac dowolnej kolejnosci argumentow
```

```python
def funkcja(*args): # ZAAWANSOWANE!!!!
    print(args)

funkcja(1,2,2,2,2,2,2,2,2)
```

```python
def funkcja(**kwargs): # ZAAWANSOWANE!!!!
    print(kwargs)
    
funkcja(x=1,y=2)
```
